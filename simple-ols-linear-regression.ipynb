```python
df = pd.read_csv("/kaggle/input/indias-population-simple-linear-regression/population.csv")
```

```python
X = df[['Year']].values
y = df['Population'].values
```

```python
test_size = 0.2
n_samples = len(X)
n_test = int(test_size * n_samples)
indices = np.arange(n_samples)
np.random.shuffle(indices)
train_indices = indices[:-n_test]
test_indices = indices[-n_test:]

```

```python
X_train, X_test = X[train_indices], X[test_indices]
y_train, y_test = y[train_indices], y[test_indices]
```

```python
class maaz_linearRegression:
    def __init__(self):
        self.theta_one = None
        self.theta_zero = None
        
        
    def fit(self, X_train,y_train):
        num = 0
        den = 0
        
        for i in range(X_train.shape[0]):
            num = num + ((X_train[i] - X_train.mean())*(y_train[i] - y_train.mean()))
            den = den + ( X_train[i] - X_train.mean() )*( X_train[i] - X_train.mean())
             
        self.theta_one = num/den
        self.theta_zero = y_train.mean() - (self.theta_one * X_train.mean())
        print(self.theta_one)
        print(self.theta_zero)
                                                         
    def predict(self,X_test):
        return self.theta_one * X_test + self.theta_zero
```

```python
model = maaz_linearRegression()
model.fit(X_train,y_train)
```
